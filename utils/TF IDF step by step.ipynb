```python
# packages
import numpy as np
import pandas as pd

# constants
SEPARATOR = " "
```

Inputs

```python
docA = "this is not not a test"
docB = "not another test" 
```

Split input using space as separator

```python
bowA = docA.split(SEPARATOR)
bowB = docB.split(SEPARATOR)
```

Store the vocabulary

```python
wordSet = set(bowA).union(set(bowB))
```

Count frecuencies

```python
wordDictA = dict.fromkeys(wordSet, 0) 
wordDictB = dict.fromkeys(wordSet, 0) 
```

```python
for word in bowA:
    wordDictA[word]+=1
    
for word in bowB:
    wordDictB[word]+=1
```

Create DataFrame with word counts

```python
pd.DataFrame([wordDictA, wordDictB])
```

Function to compute the TF

```python
def computeTF(wordDict, bow):
    tfDict = {}
    bowCount = len(bow)
    for word, count in wordDict.items():
        tfDict[word] = count/float(bowCount)
    return tfDict
```

```python
tfBowA = computeTF(wordDictA, bowA)
tfBowB = computeTF(wordDictB, bowB)
```

Function to compute IDF

```python
def computeIDF(docList):
    import math
    idfDict = {}
    N = len(docList)
    
    # create dict using the first element
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        # count in how many docs appears each term
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    # idfDict contains the number of docs in wich term appears
    for word, val in idfDict.items():
        # compute the idf
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict    
```

```python
idfs = computeIDF([wordDictA, wordDictB])
```

TF-IDF is simply the multiplication of TF and IDF values

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val * idfs[word]
    return tfidf
```

```python
tfidfBowA = computeTFIDF(tfBowA, idfs)
tfidfBowB = computeTFIDF(tfBowB, idfs)
```

Using some function to normalize vectors

```python
def get_norm(w):
  return np.linalg.norm(w)
```

```python
df = pd.DataFrame([tfidfBowA, tfidfBowB])
# get norms by row
norms = df.apply(get_norm, axis=1)
# divide each value by its norm
df.div(norms, axis=0)
```
